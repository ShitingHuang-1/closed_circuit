```python
import numpy as np
import scipy as sp
import sympy
import math
import matplotlib.pyplot as plt
from heart import *
```

```python
T=0.8
delta_t=0.001
t=np.arange(0,14,delta_t)
vt=(-69.51672*np.cos(7.85398*t)+180)
dvdt=546*np.sin((2*math.pi/T)*t)
dqdt=4288.2739*np.cos(7.85398*t)
```

```python
a=heart(0.269*T,0.452*T,1.32,27.4,0.6,0.04,20*10**(-9),55,T)
#tau1,tau2,m1,m2,Emax,Emin,Ks,V0
```

```python
pt=[]
Et=[]
Rs=[]
for i in range(len(t)):
    pt.append(a.p(vt[i],t[i]))
    Et.append(a.Et(t[i]))
    Rs.append(a.Rs(vt[i],t[i]))
#pt=pt[400:1400]
```

```python
fig,axs = plt.subplots(5,1,sharex=True)
axs[0].plot(t,pt)
axs[1].plot(t, Et)
axs[2].plot(t, vt)
axs[3].plot(t,dvdt)
axs[4].plot(t,dqdt)
#plt.xlim(4, 7)  # setting x scale
#plt.ylim(4, 6)  
plt.show()
```

```python
import numpy as np
import scipy as sp
import sympy
import math
import matplotlib.pyplot as plt
import valve_q
```
