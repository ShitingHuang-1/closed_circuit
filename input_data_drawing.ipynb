```python
import numpy as np
data = np.load('sobol_pul_circ.npz')
print(data.files)
for array_name in data.files:
    print(f"Array name: {array_name}")
    print(data[array_name])
```

```python
import matplotlib.pyplot as plt
t = np.arange(0,0.8,0.01)
fig,axs = plt.subplots(4, 2, figsize=(16,10),sharex=True)
label = ['Za_p','R_p','C_p','Emax_rv','Emin_rv']
axs[0,0].plot(t, data['S1_plv'], label =label)
axs[0,1].plot(t,data['ST_plv'],label=label)


axs[1,0].plot(t, data['S1_pao'], label =label)
axs[1,1].plot(t, data['ST_pao'], label =label)


axs[2,0].plot(t, data['S1_qav'], label =label)
axs[2,1].plot(t, data['ST_qav'], label =label)

axs[3,0].plot(t, data['S1_qpv'], label =label)
axs[3,1].plot(t, data['ST_qpv'], label =label)


axs[3,1].set_xlabel('Time(s)')
axs[3,0].set_xlabel('Time(s)')
for row in axs:
    for ax in row:
        ax.set_ylim(-0.1, 1.1)
plt.savefig('11.svg')
plt.show()
```
